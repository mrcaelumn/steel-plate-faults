```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
# load dataframe

df = pd.read_csv("datasets/stell-faults.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
# function for converting value for each faults columns to a single column as label
def gets_faults_column(row):
    if row['Pastry'] == 1:
        return "Pastry"
    elif row['Z_Scratch'] == 1:
        return "Z_Scratch"
    elif row['K_Scatch'] == 1:
        return "K_Scatch"
    elif row['Stains'] == 1:
        return "Stains"
    elif row['Dirtiness'] == 1:
        return "Dirtiness"
    elif row['Bumps'] == 1:
        return "Bumps"
    elif row['Other_Faults'] == 1:
        return "Other_Faults"
    
    return "Other_Faults"
```

```python
# checking null value in every columns
df.isnull().sum()
```

```python
df['Faults'] = df.apply(gets_faults_column, axis=1)
```

```python
df.head()
```

```python
# checking null value in every columns
df.isnull().sum()
```

```python
# drop the faults columns because we already replace with the new one
df = df.drop(['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains', 'Dirtiness', 'Bumps', 'Other_Faults'], axis = 1)

df.head()
```

```python
# split dataset into train and test dataset with ratio 7:3
# get labels
y = df.pop('Faults')
print(y)
X = df
X_train, y_train, X_test, y_test = train_test_split(X,y, test_size=0.3, random_state=43)
```
